```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_squared_error
```

```python
x = np.arange(0,500)
x[:10]
```

# 1.now make a fake  $ f(x) = 7x + 12 $ as goal for our Machine learning 

```python
y = 7*x + 12 + np.random.uniform(-250 ,250 , (500,))
```

```python
y[:10]
```

```python
plt.scatter(x[::10] , y[::10] , marker=".");
```

```python

m = y.mean()
m
```

```python

plt.scatter(x[::10] , y[::10] , marker=".")
plt.plot([0,500], [m, m], 'r');
```

## 2. now make better formula to describe data x -> y and show how dummy is mean model 

```python
x[:10]
```

```python
f = x.reshape(-1,1)
f[:10]
```

```python
model = LinearRegression()
```

```python
model.fit(x.reshape(-1,1) , y.reshape(-1,1))
```

### Now findout Coefficient of X

```python
print(model.coef_) 
```

### Now findout intercept 

```python
print(model.intercept_)
```

```python
p = model.predict(x.reshape(-1,1))
```

```python
p[:10]
```

```python
plt.scatter(x[::10] , y[::10] , marker=".")
plt.plot([0,500], [m, m], 'r');
plt.plot(x[::10],p[::10] , 'b'); 
```

### 3. Evaluate Regression Model with R2 

```python
model.score(x.reshape(-1,1) , y.reshape(-1,1))
```

### Mean Square Error 

```python
mean_squared_error(y,p)
```
